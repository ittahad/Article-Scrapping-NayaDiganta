```
from requests import get
import re
url = 'http://www.dailynayadiganta.com/archive/2019-06-05'
response = get(url)

from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)

article_containers = html_soup.find_all('li')
fileNo = 1
for cont in article_containers:
  if (cont.find('a') is not None):
      if(cont.a.h1 is not None):
        cont2 = cont.a.find('h1', class_='title')
#         print(cont2.text)
#         print(cont.a['href'])
        urlDeep = cont.a['href']
        responseDeep = get(urlDeep)
        html_soup_deep = BeautifulSoup(responseDeep.text, 'html.parser')
        contDeep = html_soup_deep.find('div', class_="news-content")
        
        text = ""
        for rows in contDeep:

          text += str(rows.string)
          finaltext = re.sub(r'[A-Za-z0-9\n]', '', text)
          text = finaltext
        
        print(text)
        print("\n")
       
        if(text != ""):
       
          with open('/content/gdrive/My Drive/News/june05.' + str(fileNo) + '.txt', 'w') as f:
            f.write(text)

          fileNo += 1
      

```
